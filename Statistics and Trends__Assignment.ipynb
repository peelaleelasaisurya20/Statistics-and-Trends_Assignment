```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('Top_Anime_data.csv')
df.head()
```

```python
df.rename(columns={'English': 'Anime_Name'}, inplace=True)
required_columns=['Anime_Name','Episodes', 'Score', 'Popularity', 'Rank']
new_df=df[required_columns]
new_df.head()
```

```python
def plot_Score_histogram(new_df):
    # Extract the 'Score' values
    Values = new_df['Score']
    
    # Plotting the histogram
    plt.figure(figsize=(8,6))
    plt.hist(Values, bins=10, color='skyblue', edgecolor='black')
    
    # Adding labels and title
    plt.title('Score Distribution of Top 10 Anime')
    plt.xlabel('Score')
    plt.ylabel('Frequency')
    
    # Display the plot
    plt.show()
    return


plot_Score_histogram(new_df)
```
